```python
import time
from time import sleep
from selenium import webdriver
import csv
import pandas as pd
from datetime import datetime as dt
```

```python
def google_ads_scraper(**kwargs):
    begin_time = dt.now() #execution time starts
    for k,v in kwargs.items():
        if k=="url":
            url = v
        else:
            search_string = v
    if search_string:
            url="http://www.google.com/search?q="+search_string
    driver = webdriver.Chrome()
    driver.maximize_window()
    driver.get(url)
    print("Opening Url")
    
    product_name = {}
    product_price = {}
    product_company = {}
    delivery_or_rating = {}
    product_link = {}
    product_image = {}

    #topmost class
    res = driver.find_element_by_class_name('top-pla-group-inner')
    
    #all_in_scrapping_starts_here
    res2 = res.find_elements_by_class_name('orXoSd')
    for i in range(len(res2)):
        try:
            product_name[i] = res2[i].text.split('\n')[0]
        except Exception as e:
            product_name[i] = 'Not found'
        try:
            product_price[i] = res2[i].text.split('\n')[1]
        except Exception as e:
            product_price[i] = 'Not found'
        try:
            product_company[i] = res2[i].text.split('\n')[2]
        except Exception as e:
            product_company[i] = 'Not found'
        try:
            delivery_or_rating[i] = res2[i].text.split('\n')[3]
        except Exception as e:
            #print(e)
            delivery_or_rating[i] = 'Not found'

        
    #product_link_scrapping_starts_here
    res3 = res.find_elements_by_class_name('pla-unit.mnr-c')
    for i in range(len(res3)-1):
        tag = res3[i].find_element_by_tag_name('a') #works fine
        product_link[i] = tag.get_attribute('href') #works fine
    
    
    #product_image_scraping_starts_here
    res4 = res.find_elements_by_class_name('pla-unit.mnr-c')
    for i in range(len(res4)-1):
        try:
            image_link = res4[i].find_element_by_id("platop"+str(i)).get_attribute("src")
        except Exception as e:
            #print(e)
            continue
        product_image[i] = image_link
    
    
    #printing all the dictionaries
    #rint(len(product_name))
    #rint(len(product_price))
    #rint(len(product_company))
    #rint(len(delivery_or_rating))
    #rint(len(product_link))
    #rint(len(product_image))
    
    #printing execution time
    print('google_ad_scraper execution time is {}'.format(dt.now() - begin_time))
    
    
    #writing csv
    dicts = product_name,product_price,product_company,delivery_or_rating,product_link,product_image
    
    filename = datetime.now().strftime('scraped_data-%Y-%m-%d-%H-%M.csv') #saves_file_name_with_date_and_time_in_24_hours_format
    
    with open(filename, 'w',encoding="utf-8") as ofile:
        writer = csv.writer(ofile, delimiter=',')
        writer.writerow(['product_name','product_price','product_company','delivery_or_rating','product_link','product_image'])
        for key in dicts[0].keys():
            writer.writerow([d[key] for d in dicts])
    
    
    return pd.read_csv('scraped_data.csv')
```
